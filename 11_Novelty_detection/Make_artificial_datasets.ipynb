```python
# 1 center
centers = [[0, 0], [1, 1]]
normal, _ = make_blobs(n_samples=980, centers=centers, cluster_std=0.4, random_state=123)
# abnormal, _ = make_blobs(n_samples=20, cluster_std=0.1, centers=[[1.5, 1.5], [1.3, -1.3], [-1.0, -1.5], [-1.0, 1.3]])
np.random.seed(584)
abnormal = np.random.uniform(low=-3, high=3, size=(50, 2))
plt.scatter(normal[:, 0], normal[:, 1], marker='x', color='b', label='normal point')
plt.scatter(abnormal[:, 0], abnormal[:, 1], marker='x', color='r', label='abnormal point')
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.title('Artificial dataset')
plt.show()
```

```python
_normal = np.c_[normal, np.zeros(len(normal))]
_abnormal = np.c_[abnormal, np.ones(len(abnormal))]
_data = np.concatenate((_normal, _abnormal), axis=0)

data = pd.DataFrame(data=_data, columns=['X1', 'X2', 'class'])
```
